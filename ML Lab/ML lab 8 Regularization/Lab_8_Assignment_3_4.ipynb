```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BDSET /Lab/Lab_8/dataset 2.csv')
df.head()
```

```python
plt.scatter(df.X1,df.X2)
```

```python
x = df[["X1"]].to_numpy()
y = df[["X2"]].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state = 2)
```

```python
plt.scatter(x_train, y_train, color = 'blue')
plt.scatter(x_test, y_test, color = 'red')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
print(mean_squared_error(y_test, y_pred))
```

```python
plt.plot(x_train, lr.predict(x_train), color = 'r')
plt.plot(x, y, "b.")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

#Ridge

```python
from sklearn import linear_model
ridge = linear_model.Ridge(alpha = .8)
ridge.fit(x_train, y_train)
y_pred_ridge = ridge.predict(x_test)
print(mean_squared_error(y_test, y_pred_ridge))
```

```python
plt.plot(x_train, ridge.predict(x_train), color="r")
plt.plot(x, y, "b.")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

#Ridge KFold

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=3, shuffle = True, random_state =2)
for train_index, test_index in kf.split(x):
  # print("Train:", train_index, "Validation:",test_index)
  x_train, x_test = x[train_index], x[test_index]
  y_train, y_test = y[train_index], y[test_index]

  from sklearn import linear_model
  ridge = linear_model.Ridge(alpha = .8)
  ridge.fit(x_train, y_train)
  y_pred_ridge = ridge.predict(x_test)
  print(mean_squared_error(y_test, y_pred_ridge))

  plt.plot(x_train, ridge.predict(x_train), color="r")
  plt.plot(x, y, "b.")
  plt.xlabel("X")
  plt.ylabel("Y")
plt.show()
```

```python
np.linspace(0.1, 2, 20)
```

```python
alpha = np.linspace(0.1, 5, 20)

for i in (alpha):
  kf = KFold(n_splits=3, shuffle = True, random_state =2)
  for train_index, test_index in kf.split(x):
    # print("Train:", train_index, "Validation:",test_index)
    x_train, x_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]

    from sklearn import linear_model
    ridge = linear_model.Ridge(alpha = i)
    ridge.fit(x_train, y_train)
    y_pred_ridge = ridge.predict(x_test)
    print(mean_squared_error(y_test, y_pred_ridge))

    plt.plot(x_train, ridge.predict(x_train), color="r")
    plt.plot(x, y, "b.")
    plt.xlabel("X")
    plt.ylabel("Y")
  plt.show()
```

#Lasso

```python
lasso = linear_model.Lasso(alpha=.8)
lasso.fit(x_train, y_train)
y_pred_lasso = lasso.predict(x_test)
print(mean_squared_error(y_test, y_pred_lasso))
```

```python
plt.plot(x_train, lasso.predict(x_train), color="r")
plt.plot(x, y, "b.")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
