#The gaussian function

```python


import math

def gaussiankernal(xt,yt):
  
  xt=xt
  yt=yt
  sigma_x = np.std(xt)
  sigma_y = np.std(yt)
  ro= np.corrcoef(xt,yt)[0,1]
  z = math.exp((-1)/(2*(1-ro**2))*((xt**2/sigma_x**2)+(yt**2/sigma_y**2)-((2*ro*xt*yt)/(sigma_x*sigma_y))))
  return z
```
